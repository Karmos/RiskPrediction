## Descrizione del dataset
Il dataset presenta le seguenti informazioni:
- addr_state: Lo stato fornito dal mutuatario nella domanda di prestito
- annual_inc: Il reddito annuo auto-dichiarato fornito dal mutuatario durante la registrazione
- annual_inc_joint: Il reddito annuo complessivo auto-dichiarato fornito dai co-mutuatari durante la registrazione
- application_type: Indica se il prestito è una domanda individuale o una domanda congiunta con due co-mutuatari
- collection_recovery_fee: Commissione di recupero post-incasso
- collections_12_mths_ex_med: Numero di collezioni negli ultimi 12 mesi escluse quelle mediche
- delinq_2yrs: Numero di incidenti di insolvenza di oltre 30 giorni nel file del credito del mutuatario negli ultimi 2 anni
- desc: Descrizione del prestito fornita dal mutuatario
- dti: Un rapporto calcolato utilizzando i pagamenti mensili totali del mutuatario rispetto agli obblighi totali del debito
- dti_joint: Un rapporto calcolato utilizzando i pagamenti mensili totali dei co-mutuatari rispetto agli obblighi totali del debito
- earliest_cr_line: Il mese in cui è stata aperta la linea di credito più antica del mutuatario
- emp_length: Anni di esperienza lavorativa. I valori possibili vanno da 0 a 10, dove 0 significa meno di un anno di esperienza
- emp_title: Titolo del lavoro fornito dal mutuatario al momento della domanda di prestito
- fico_range_high: Il limite superiore della fascia FICO del mutuatario al momento della concessione del prestito
- fico_range_low: Il limite inferiore della fascia FICO del mutuatario al momento della concessione del prestito
- funded_amnt: L'importo totale impegnato per quel prestito in quel momento
- funded_amnt_inv: L'importo totale impegnato dagli investitori per quel prestito in quel momento
- grade: Grado del prestito assegnato da LC
- home_ownership: Stato di proprietà dell'immobile fornito dal mutuatario durante la registrazione (valori: AFFITTO, PROPRIETARIO, MUTUO)
- id: ID unico assegnato da LC per l'elenco del prestito
- initial_list_status: Lo stato iniziale dell'elenco del prestito. I valori possibili sono: W, F
- inq_last_6mths: Numero di richieste di credito negli ultimi 6 mesi (escludendo le richieste per auto e mutuo)
- installment: Pagamento mensile dovuto dal mutuatario se il prestito viene concesso
- int_rate: Tasso di interesse del prestito
- is_inc_v: Indica se il reddito è stato verificato da LC, non verificato, o se è stata verificata la fonte di reddito
- issue_d: Mese in cui il prestito è stato finanziato
- last_credit_pull_d: Il mese più recente in cui LC ha effettuato una verifica del credito per questo prestito
- last_fico_range_high: Il limite superiore della fascia FICO dell'ultima verifica del mutuatario
- last_fico_range_low: Il limite inferiore della fascia FICO dell'ultima verifica del mutuatario
- last_pymnt_amnt: Ultimo importo totale ricevuto come pagamento
- last_pymnt_d: Ultimo mese in cui è stato ricevuto un pagamento
- loan_amnt: Importo del prestito richiesto dal mutuatario
- loan_status: Stato attuale del prestito
- member_id: ID unico assegnato dal LC per il membro mutuatario
- mths_since_last_delinq: Numero di mesi dall'ultima insolvenza del mutuatario
- mths_since_last_major_derog: Mesi dall'ultimo rating negativo di 90 giorni o peggio
- mths_since_last_record: Numero di mesi dall'ultimo record pubblico
- next_pymnt_d: Data del prossimo pagamento programmato
- open_acc: Numero di linee di credito aperte nel file del credito del mutuatario
- out_prncp: Principale residuo non ancora pagato per l'importo totale finanziato
- out_prncp_inv: Principale residuo non ancora pagato per la porzione dell'importo totale finanziato dagli investitori
- policy_code: Codice della politica pubblicamente disponibile = 1, nuovi prodotti non pubblicamente disponibili = 2
- pub_rec: Numero di record pubblici negativi
- purpose: Categoria fornita dal mutuatario per la richiesta di prestito
- pymnt_plan: Indica se è stato messo in atto un piano di pagamento per il prestito
- recoveries: Recupero lordo post-incasso
- revol_bal: Bilancio totale del credito revolving
- revol_util: Tasso di utilizzo della linea di credito revolving, ovvero l'importo di credito che il mutuatario sta utilizzando rispetto a quello disponibile
- sub_grade: Sottogrado del prestito assegnato da LC
- term: Numero di pagamenti previsti per il prestito (36 o 60 mesi)
- title: Titolo del prestito fornito dal mutuatario
- total_acc: Numero totale di linee di credito nel file del credito del mutuatario
- total_pymnt: Pagamenti ricevuti fino ad oggi per l'importo totale finanziato
- total_pymnt_inv: Pagamenti ricevuti fino ad oggi per la porzione dell'importo totale finanziato dagli investitori
- total_rec_int: Interessi ricevuti fino ad oggi
- total_rec_late_fee: Commissioni per ritardo ricevute fino ad oggi
- total_rec_prncp: Capitale ricevuto fino ad oggi
- url: URL della pagina LC con i dati dell'elenco
- verified_status_joint: Indica se il reddito congiunto dei co-mutuatari è stato verificato da LC, non verificato, o se è stata verificata la fonte di reddito
- zip_code: I primi 3 numeri del CAP forniti dal mutuatario nella domanda di prestito
- open_acc_6m: Numero di operazioni aperte negli ultimi 6 mesi
- open_il_6m: Numero di operazioni di prestiti a rate attive
- open_il_12m: Numero di conti di prestito a rate aperti negli ultimi 12 mesi
- open_il_24m: Numero di conti di prestito a rate aperti negli ultimi 24 mesi
- mths_since_rcnt_il: Mesi dall'ultimo conto di prestito a rate aperto
- total_bal_il: Bilancio totale attuale di tutti i conti di prestito a rate
- il_util: Rapporto tra il saldo attuale totale e il limite di credito/credito massimo di tutti i conti a rate
- open_rv_12m: Numero di operazioni di credito revolving aperte negli ultimi 12 mesi
- open_rv_24m: Numero di operazioni di credito revolving aperte negli ultimi 24 mesi
- max_bal_bc: Bilancio massimo attuale dovuto su tutti i conti di credito revolving
- all_util: Rapporto saldo/limite di credito su tutti i conti
- total_rev_hi_lim: Limite totale di credito revolving
- inq_fi: Numero di richieste di finanziamento personale
- total_cu_tl: Numero totale di operazioni finanziarie
- inq_last_12m: Numero di richieste di credito negli ultimi 12 mesi
- acc_now_delinq: Numero di conti in cui il mutuatario è attualmente in insolvenza
- tot_coll_amt: Importo totale delle collezioni mai dovute
- tot_cur_bal: Saldo totale attuale di tutti i conti

Osserviamo adesso le proporzioni dei possibili stati del prestito in riferimento alla durata dell'impiego

```python
# Define the desired order for emp_length groups
emp_length_groups = ['<1', '1-4', '5-8', '>8']

# Create a copy of the DataFrame to avoid modifying the original
df_copy = df.copy()

# Define a function to group emp_length into the desired categories
def group_emp_length(emp_length):
    if emp_length == '< 1 year':
        return '<1'
    elif emp_length in ['1 year', '2 years', '3 years', '4 years']:
        return '1-4'
    elif emp_length in ['5 years', '6 years', '7 years', '8 years']:
        return '5-8'
    else:
        return '>8'

# Apply the grouping function to create a new column in df_copy
df_copy['emp_length_group'] = df_copy['emp_length'].apply(group_emp_length)

# Convert the new 'emp_length_group' column to a categorical type with the specified order
df_copy['emp_length_group'] = pd.Categorical(df_copy['emp_length_group'], categories=emp_length_groups, ordered=True)

# Set up the figure
plt.figure(figsize=(10, 6))

# Create the stacked histogram with the grouped emp_length data
sns.histplot(data=df_copy, x="emp_length_group", hue="loan_status", multiple="stack", palette='bright')


# Add title and axis labels
plt.title("Relationship between Employment Length and Loan Status")
plt.xlabel("Employment Length")
plt.ylabel("Count")

# Show the plot
plt.show()
```

Notiamo come per gli stati dove non si rilevano criticità (current, fully paid) e quelli con piccoli ritardi nei pagamenti (In grace period, Late (16-30 days) si riscontra un reddito mediano annuo superiore ai $60000, mentre a ritardi maggiori (31-120 days) e casi di insolvenza (default e charged-off) corrispondono redditi  inferiori.

```python
# Count the number of loans per term
loans_per_term = df['term'].value_counts().reset_index()

# Group by term and loan_status to calculate the count and percentage
count_data = df.groupby(['term', 'loan_status']).size().reset_index(name='count')
count_data['percentage'] = count_data['count'] / count_data.groupby('term')['count'].transform('sum') * 100

# Set up the figure with a larger size
fig = plt.figure(figsize=(16, 6))

# Create grid spec for flexible layout
gs = gridspec.GridSpec(1, 2, width_ratios=[1, 4])

# First subplot: Barplot showing the number of loans per term
ax1 = plt.subplot(gs[0])
sns.barplot(x='term', y='count', data=loans_per_term, ax=ax1, width=0.4, palette='bright')
ax1.set_title('Loans per Term', fontsize=16, weight='bold', pad=20)
ax1.set_xlabel('Loan Term', fontsize=12, weight='bold', labelpad=10)
ax1.set_ylabel('Number of Loans', fontsize=12, weight='bold', labelpad=10)

# Second subplot: Barplot showing loan status percentages by loan term
ax2 = plt.subplot(gs[1])
sns.barplot(x='term', y='percentage', hue='loan_status', data=count_data, ax=ax2, width=0.8, palette='bright')

# Add percentage labels on top of bars
for p in ax2.patches:
    height = p.get_height()
    if height > 0:
        ax2.text(
            p.get_x() + p.get_width() / 2.,
            height,
            f'{height:.2f}%',
            ha='center',
            va='bottom',
            fontsize=10,
            weight='bold'
        )

# Customize second subplot titles and labels
ax2.set_title('Loan Status by Loan Term', fontsize=16, weight='bold', pad=20)
ax2.set_xlabel('Loan Term', fontsize=12, weight='bold', labelpad=10)
ax2.set_ylabel('Percentage per Loan Status (%)', fontsize=12, weight='bold', labelpad=10)

# Improve the legend to make it more readable
ax2.legend(title='Loan Status', fontsize=10, loc='upper left')

# Adjust layout to avoid overlap and ensure everything fits
plt.tight_layout()

# Display the plot
plt.show()
```

Si osserva che la maggior parte dei prestiti ha una durata di 36 mesi. Per entrambe le tipologie di prestiti, la maggior parte risulta ancora in corso. È interessante notare che, per i prestiti a lunga durata, la percentuale di prestiti in corso è più alta, mentre quella di prestiti risarciti è più bassa, il che non sorprende data la maggiore durata. Le proporzioni per le altre categorie di prestiti sono simili.

```python
def sub_grades_encoding(df):
    # Define a dictionary mapping letter grades to numeric values
    grade_map = {
        'A': 7,
        'B': 6,
        'C': 5,
        'D': 4,
        'E': 3,
        'F': 2,
        'G': 1
    }

    # Vectorized operation for letter grade encoding
    grade_vals = df['sub_grade'].str.extract('([A-G])')[0].map(grade_map)

    # Vectorized operation for numeric adjustment based on 1-5
    numeric_adjustments = {
        '1': 0.8,
        '2': 0.6,
        '3': 0.4,
        '4': 0.2,
        '5': 0.0
    }

    # Extract the numeric suffix (1, 2, 3, etc.) and apply the corresponding adjustment
    numeric_vals = df['sub_grade'].str.extract('([1-5])')[0].map(numeric_adjustments).fillna(0)

    # Combine the grade and numeric adjustments
    df['sub_grade'] = grade_vals + numeric_vals

    return df

# verification status encoding
def verification_status(x):
    # Return 0 for 'Not Verified' and 1 for any other status
    return 0 if x == 'Not Verified' else 1
```

```python
# Applying the vectorized function
df = sub_grades_encoding(df)
```

## Gestione dei valori mancanti

Si verifica se ci sono valori mancanti (NaN) nel dataFrame e nel caso si applica l'algoritmo di imputazione iterativa per riempirli, utilizzando un modello che considera la correlazione tra le diverse colonne.

```python
# # Importing necessary libraries
from sklearn.experimental import enable_iterative_imputer #to enable iterativeImputer
from sklearn.impute import IterativeImputer

if df.isnull().any().any():
    # Initialize the IterativeImputer with optimized settings
    imputer = IterativeImputer(max_iter=10, random_state=42, n_nearest_features=5)
    # Apply the IterativeImputer to the DataFrame
    imputed_data = imputer.fit_transform(df)
    # Convert the imputed data back to a DataFrame
    imputed_df = pd.DataFrame(imputed_data, columns=df.columns)
```
